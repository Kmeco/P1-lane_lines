# **Finding Lane Lines on the Road.**

## Import Packages

```python
#importing some useful packages
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
%matplotlib inline
```

## Defining the Functions

```python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines_new(img, lines, color=[255, 0, 0], thickness=5):
    """

    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    # defining the empty arrays which will represent the endpoints of the two lines (positive and negative gradient)
    j=0
    x_pos=np.empty(0)
    y_pos=np.empty(0)
    x_neg=np.empty(0)
    y_neg=np.empty(0)
    
    # computing the gradient of each line in the lines output from HoughLinesP and 
    # delete those which have the gradient btw -0.5 and 0.5
    for i in lines: 
        grad=(i[0][3]-i[0][1])/(i[0][2]-i[0][0])
        if -0.5<grad<0.5:
            lines=np.delete(lines, j, 0)
            j-=1
        j+=1
    # devide the endpoints of positive and negative lines into separate 
    # arrays containing only x and only y coordinates.
        if grad>0:
            x_pos=np.append(x_pos, [i[0][0], i[0][2]])
            y_pos=np.append(y_pos, [i[0][1], i[0][3]])
            
        if grad<0:
            x_neg=np.append(x_neg, [i[0][0], i[0][2]])
            y_neg=np.append(y_neg, [i[0][1], i[0][3]])
    # deleting any outliers in the dataset of endpoints 
    for v in range(2):
        z_pos=np.polyfit(x_pos, y_pos, 1)   #fiting a line
        z_neg=np.polyfit(x_neg, y_neg, 1)

        p_pos=np.poly1d(z_pos)
        p_neg=np.poly1d(z_neg)

        pos_diff=np.empty(0) 
        f=0
        for i in x_pos:
            pos_diff=np.append(pos_diff, abs(p_pos(i)-y_pos[f]))    #calculating a difference btw the real and predicted value 
            f+=1

        size=int(np.size(pos_diff)*0.1)     #defininf 10% of the dataset
        n=0
        while n<size:                            #deleting 10% of the dataset with the highest residual errror
            index=np.argmax(pos_diff)
            x_pos, y_pos= np.delete(x_pos, index), np.delete(y_pos, index)  
            pos_diff=np.delete(pos_diff, index)
            n+=1
            
        # if uncommented this can be used to visualise the dataset and any outliers
        #plt.figure()
        #plt.scatter(x_pos, y_pos)
        #plt.show()

        z_pos=np.polyfit(x_pos, y_pos, 1)   # fitting a new line for the cleaner dataset and repeating the process 2 more times
        z_neg=np.polyfit(x_neg, y_neg, 1)
    
    y1=320                  # this is arbitrarry depending on the horizont in our video
    y2=np.size(image, 1)    # the lower edge of our video
    
    x1_pos=int((y1-z_pos[1])/z_pos[0])      # extrapolating and calculating the endpoints of our final lines
    x2_pos=int((y2-z_pos[1])/z_pos[0])
    
    x1_neg=int((y1-z_neg[1])/z_neg[0])
    x2_neg=int((y2-z_neg[1])/z_neg[0])
 
    cv2.line(img, (x1_pos, y1), (x2_pos, y2), color, thickness)    # displaying the lines
    cv2.line(img, (x1_neg, y1), (x2_neg, y2), color, thickness)

def draw_lines(img, lines, color=[255, 0, 0], thickness=3):  
    """
    this is an alternative to draw_lines_new which desplays 
    all Hough Lines. 
    If you wish to put run this, the argument new_old in Hough_lines 
    function needs to be set to equal 0."""
    
    j=0
    for i in lines: 
        grad=(i[0][3]-i[0][1])/(i[0][2]-i[0][0])          # removing any horizontal lines with gradient btw -0.5 and 0.5
        if -0.5<grad<0.5:
            lines=np.delete(lines, j, 0)
            j-=1
        j+=1
        
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)
    
            

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap, new_old=1):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    
    For new_old=0 - prints all the Hough Lines in the region of interest.
    For new_old=1 - prints an averaged and extrapolated line for each lane line. 
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)

    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    if new_old == 0:
        draw=draw_lines
    else:
        draw=draw_lines_new
        
    draw(line_img, lines)
    return line_img

def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    
    return cv2.addWeighted(initial_img, α, img, β, λ)
```

## Develop a pipeline for a set of test images 

```python
import os
img_list=os.listdir("test_images/")

for i in img_list:     #runs through all images in the test_images folder
    
    pic="test_images/"+i
    
    image = mpimg.imread(pic)
    #convert the image into grayscale
    gray=grayscale(image)
    
    #applying gaussian blur before running the canny edge function
    kernel_size=5
    blur_gray=gaussian_blur(gray, kernel_size)

    
    low_threshold = 50
    high_threshold = 150
    canny_img=canny(blur_gray, low_threshold, high_threshold)
    
    #uncomment the code below to display the canny edge output
    #plt.figure()
    #plt.imshow(canny_img, cmap='Greys_r')
    
    #applying the mask to delete any irrelevant edges
    mask = np.zeros_like(canny_img)   
    ignore_mask_color = 255
    
    imshape = image.shape
    vertices = np.array([[(30,imshape[0]),(415, 320), (505, 320), (imshape[1]-20,imshape[0])]], dtype=np.int32)
    
    masked_canny_img=region_of_interest(canny_img, vertices)
    
    rho = 2 # distance resolution in pixels of the Hough grid
    theta = np.pi/180 # angular resolution in radians of the Hough grid
    threshold = 5    # minimum number of votes (intersections in Hough grid cell)
    min_line_len = 30 #minimum number of pixels making up a line
    max_line_gap = 20   # maximum gap in pixels between connectable line segments
    
    
    hough_img=hough_lines(masked_canny_img, rho, theta, threshold, min_line_len, max_line_gap)
    
    result=weighted_img(hough_img, image, α=0.8, β=1., λ=0.)
    
    plt.figure()
    plt.imshow(result, cmap='Greys_r')
```
